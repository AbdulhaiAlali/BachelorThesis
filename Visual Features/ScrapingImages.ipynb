```python
import pandas as pd
import os

# Read the second dataset
df2 = pd.read_csv('Datasets/raw/CarNamesClean.csv')

# Path to the folder containing CSV files
folder_path = 'Datasets/CleanedHisPrices/'

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        # Read the CSV file into a DataFrame
        stock_name = os.path.splitext(filename)[0]
        df = pd.read_csv(os.path.join(folder_path, filename))
        # Add the Stock name as a column
        #df['Stock'] = stock_name
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
df1 = pd.concat(dfs, ignore_index=True)

# Merge the two DataFrames based on 'Date' and 'Stock' columns
merged_df = pd.merge(df2, df1[['Date', 'Stock', 'Price']], on=['Date', 'Stock'], how='left')

# For missing prices, find the closest available date and use its price
missing_prices = merged_df['Price'].isnull()
if any(missing_prices):
    closest_dates = []
    for idx, row in merged_df[missing_prices].iterrows():
        closest_date = df1[(df1['Stock'] == row['Stock']) & (df1['Date'] <= row['Date'])]['Date'].max()
        closest_price = df1.loc[(df1['Stock'] == row['Stock']) & (df1['Date'] == closest_date), 'Price'].values
        if closest_price:
            merged_df.at[idx, 'Price'] = closest_price[0]
            closest_dates.append(closest_date)
        else:
            closest_dates.append(None)
    # Add the new column 'Closest Date' to the DataFrame
    merged_df.loc[missing_prices, 'Closest Date'] = closest_dates

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('Datasets/raw/Full.csv', index=False)

print("Merged DataFrame saved to 'Full.csv'")
```

```python
import os
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd

def make_search_query(full_name):
    return full_name.replace(' ', '+')

def save_image(image_url, folder_path, image_name):
    try:
        image_path = os.path.join(folder_path, image_name)
        with open(image_path, 'wb') as f:
            f.write(requests.get(image_url).content)
    except FileNotFoundError as e:
        print(f"Skipping image '{image_name}' due to error: {e}")

def scrape_images(car_data, save_folder):
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    saved_count = 0
    total_count = len(car_data)
    
    for index, row in car_data.iterrows():
        query = make_search_query(row['CarName'])
        response = requests.get(f"https://www.google.com/search?q={query}&tbm=isch")
        soup = BeautifulSoup(response.text, 'html.parser')
        images = soup.find_all('img')

        if images:
            image_url = images[1]['src']
            image_name = f"{row['CarName']}.jpg"
            save_image(image_url, save_folder, image_name)
            car_data.at[index, 'ImageName'] = image_name
            saved_count += 1
        else:
            car_data.at[index, 'ImageName'] = "Not Found"

        print(f"Saved {saved_count}/{total_count} images so far.")
        
    car_data.to_csv('Datasets/raw/updated_car_data.csv', index=False)

# Read CSV
car_data = pd.read_csv('Datasets/raw/FullData.csv')

# Define the folder to save images
save_folder = 'images/'
os.makedirs(save_folder, exist_ok=True)  # Create the directory if it doesn't exist

# Scraping Images and Update CSV
scrape_images(car_data, save_folder)
```
