```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('Datasets/HOG/FullHOG.csv')

# Ensure 'Price' column is numerical
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Extract feature columns (assuming they start with 'Feature_')
feature_columns = df.columns[df.columns.str.startswith('PC')].tolist()

# Calculate the mean correlation of features with 'Price' for each 'Stock'
correlations = df.groupby('Stock')[feature_columns].corrwith(df['Price']).mean(axis=1)

# Get the top 10 car companies with the highest mean correlation to 'Price'
top_companies = correlations.abs().sort_values(ascending=False).head(30)

# Plot these correlations using a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x=top_companies.values, y=top_companies.index, palette='coolwarm')
plt.title('Top 10 Car Companies Most Correlated with Price')
plt.xlabel('Mean Correlation Coefficient')
plt.ylabel('Car Companies')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

# Load the dataset
df = pd.read_csv('Datasets/HOG/FullHOG.csv')

# Verify the columns in the DataFrame
print("Columns in the dataset:", df.columns)

# Ensure the 'Price' column is present
if 'Price' not in df.columns:
    raise ValueError("The 'Price' column is not present in the dataset. Please check the CSV file.")

# Drop the 'Date' column if it exists
df = df.drop(columns=['Date'], errors='ignore')

# List of feature columns (excluding 'Price')
feature_columns = [col for col in df.columns if col.startswith('PC')]

# Calculate correlation of each feature with the stock price within each company
company_correlations = df.groupby('Stock')[feature_columns].corrwith(df['Price']).mean(axis=1)

# Calculate the mean correlation for each company
mean_company_correlations = company_correlations.groupby('Stock').mean()

print(mean_company_correlations)

# Identify the best and worst performing companies based on mean correlation
best_company = mean_company_correlations.idxmax()
worst_company = mean_company_correlations.idxmin()

# Print the mean correlations for the best and worst companies
print(f"Best Company: {best_company} with Mean Correlation: {mean_company_correlations[best_company]:.2f}")
print(f"Worst Company: {worst_company} with Mean Correlation: {mean_company_correlations[worst_company]:.2f}")

# Get representative car names for these companies
best_company_carname = df[df['Stock'] == best_company].iloc[1]['CarName']
worst_company_carname = df[df['Stock'] == worst_company].iloc[1]['CarName']

# Image file paths
image_folder = 'images'
best_company_image_path = os.path.join(image_folder, best_company_carname + '.jpg')
worst_company_image_path = os.path.join(image_folder, worst_company_carname + '.jpg')

# Function to load an image
def load_image(image_path):
    try:
        return Image.open(image_path)
    except FileNotFoundError:
        print(f"File not found: {image_path}")
        return None

# Load images
best_image = load_image(best_company_image_path)
worst_image = load_image(worst_company_image_path)

# Plot images if they exist
plt.figure(figsize=(10, 5))

if best_image:
    plt.subplot(1, 2, 1)
    plt.imshow(best_image)
    plt.title(f"Best Company: {best_company} \nMean Correlation: {mean_company_correlations[best_company]:.2f}")

if worst_image:
    plt.subplot(1, 2, 2)
    plt.imshow(worst_image)
    plt.title(f"Worst Company: {worst_company} \nMean Correlation: {mean_company_correlations[worst_company]:.2f}")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Datasets/HOG/FullHOG.csv')

# Ensure the 'Price' column is present
if 'Price' not in df.columns:
    raise ValueError("The 'Price' column is not present in the dataset. Please check the CSV file.")

# List of feature columns (excluding 'Price')
feature_columns = [col for col in df.columns if col.startswith('PC')]

# Calculate correlation of each feature with the stock price within each company
company_correlations = df.groupby('Stock')[feature_columns].corrwith(df['Price']).mean(axis=1)

# Calculate the mean correlation for each company
mean_company_correlations = company_correlations.groupby('Stock').mean()

# Identify the top performing companies based on mean correlation
top_companies = mean_company_correlations.nlargest(30)

# Plot bar graph for top companies
plt.figure(figsize=(10, 6))
top_companies.plot(kind='bar', color='skyblue')
plt.title('Companies Correlated with Price')
plt.xlabel('Company')
plt.ylabel('Mean Correlation')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

def plot_images_with_text(image_folder, image_texts):
    fig, axs = plt.subplots(1, len(image_texts), figsize=(10, 5))  # Adjust figsize as needed
    
    for i, (image_name, text) in enumerate(image_texts):
        image_path = os.path.join(image_folder, image_name)
        image = Image.open(image_path)
        axs[i].imshow(image)
        axs[i].set_title(f"{image_name}\n{text}")
        axs[i].axis('off')

    plt.show()

# Example usage
image_folder = "images"
image_texts = [("Isuzu D-Max 2014.jpg", "High Correlation \nStock Price : 12.4529915"), ("Renault Megane 2015.jpg", "Low Correlation \nStock Price : 107.73381")]  # Image names and corresponding texts

plot_images_with_text(image_folder, image_texts)
```
